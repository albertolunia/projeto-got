```python
import pandas as pd
```

```python
db = pd.read_csv("../DataSets/Game_of_Thrones.csv")
db
```

### Médias(Visualizações e avaliações)

```python
print("Visualizações:")
media_viwer = db['U.S. Viewers (Millions)'].mean()
print(f"Média de visualizações por episódio (Milhões): {media_viwer:.2f}")

print("Média de avaliações por episódio realizado por")
media_IMDb = db['IMDb Rating'].mean()
print(f"IMDb: {media_IMDb:.2f}")

media_tomato = db['Rotten Tomatoes Rating (Percentage)'].mean()
print(f"Rotten Tomatoes: {media_tomato:.2f}%")

media_metacritic = db['Metacritic Ratings'].mean()
print(f"Metacritic: {media_metacritic:.2f}")
```
